# simple linear regression
y=mx+c <br>
m=slope <br>
c=y intercept <br>

```python
# Internal Data 
x=[5,7,8,7,2,17,2,9,4,11,12,9,6]
y=[99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
print(len(x),len(y))
```

```python
# Visualize the data 
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
#Use scipy 
from scipy import stats
result=stats.linregress(x,y)
```

```python
#Slope 
print(result.slope)
#Intercept 
print(result.intercept)
```

```python
#Compute the estimated y2 
import numpy as np
x2=np.array(x)
print(type(x))
print(type(x2))
```

```python
y2=result.slope* x2 + result.intercept
```

```python
# plot the line over the data 
plt.scatter(x,y)
plt.plot(x2,y2,color='green')
```

```python
# to check the linearity 
print(result.rvalue)
```

```python
print(result.rvalue * result.rvalue)
```

### This linear regression model has a low r2 value.So,the line does not fit well with our data.
