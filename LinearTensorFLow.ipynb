# Linear Regression With Tensorflow

# Tensorflow Minimal Example

# Import the relevant libraries

```python
import numpy as np 
import matplotlib.pyplot as plt 
import tensorflow as tf
```

# Data Generation

```python
observations=1000
xs=np.random.uniform(low=-10,high=10,size=(observations,1))
zs=np.random.uniform(low=-10,high=10,size=(observations,1))
generated_inputs=np.column_stack((xs,zs))
noise=np.random.uniform(-1,1,(observations,1))
generated_targets= 2 * xs - 3 * zs + 5 + noise
```

```python
np.savez('TF4_intro',inputs=generated_inputs,targets=generated_targets)
```

# Solving with tensorflow

```python
trainind_data=np.load('TF4_intro.npz')
input_size=2
output_size=1
model=tf.keras.Sequential([
    tf.keras.layers.Dense(output_size)
])
model.compile(optimizer='sgd',loss='mean_squared_error')
model.fit(trainind_data['inputs'],trainind_data['targets'],epochs=100,verbose=2)
```

# Extract the weights and biases

```python
model.layers[0].get_weights()
```

```python
weights=model.layers[0].get_weights()[0]
biases=model.layers[0].get_weights()[1]
```

# Extract the output(make predictions)

```python
model.predict_on_batch(trainind_data['inputs']).round(1)
```

```python
trainind_data['targets'].round(1)
```

# Plotting data

```python
plt.plot(np.squeeze(model.predict_on_batch(trainind_data['inputs'])),np.squeeze(trainind_data['targets']))
plt.xlabel("Outputs")
plt.ylabel("Targets")
plt.show()
```
